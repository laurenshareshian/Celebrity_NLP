```python
#GET FIRST 1500 TMZ ARTICLE LINKS
chromedriver = "/usr/local/bin/chromedriver"
driver = webdriver.Chrome(chromedriver)
article_links = []
page_count = 1
link_count = 0
while link_count < 1500:
    try:
        url = 'http://www.tmz.com/page/'+str(page_count)+'/'
        print(url)
        driver.get(url)  
        html_source = driver.page_source  
        time.sleep(10)
        soup = BeautifulSoup(html_source,'html.parser')
        for thing in soup.find_all(class_="headline"):
            article_links.append(thing['href'])
            print(link_count, thing['href'])
            link_count = link_count + 1
    except:
        continue
    page_count = page_count + 1
    
driver.quit()
```

```python
#READ ALL OF THE TMZ ARTICLES
outfile = open('celebrity_gossip2.txt', 'w')

chromedriver = "/usr/local/bin/chromedriver"

articles = []

for count, article in enumerate(link_list):
    if count >= 500 and count < 1000:
        try:
            driver = webdriver.Chrome(chromedriver)
            url = article
            driver.get(url)  
            html_source = driver.page_source  
            time.sleep(10)
            soup = BeautifulSoup(html_source,'html.parser')

            article = ''
            for thing in soup.find_all(class_ = "all-post-body group article-content"):
                for i,thing2 in enumerate(thing.find_all('p')):
                    if i < len(thing.find_all('p'))-1:
                        article = article + thing2.text.strip()

            article = article.replace('\n', ' ')
            outfile.write(article)
            outfile.write('\n')
            
            print(count, url)
            print(article[0:100])
            
            driver.quit()
        except:
            continue


outfile.close()
```
