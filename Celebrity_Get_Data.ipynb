```python
#GET FIRST 1500 TMZ ARTICLE LINKS
chromedriver = "/usr/local/bin/chromedriver"
driver = webdriver.Chrome(chromedriver)
article_links = []
page_count = 1
link_count = 0
while link_count < 1500:
    try:
        url = 'http://www.tmz.com/page/'+str(page_count)+'/'
        print(url)
        driver.get(url)  
        html_source = driver.page_source  
        time.sleep(10)
        soup = BeautifulSoup(html_source,'html.parser')
        for thing in soup.find_all(class_="headline"):
            article_links.append(thing['href'])
            print(link_count, thing['href'])
            link_count = link_count + 1
    except:
        continue
    page_count = page_count + 1
    
driver.quit()
```

```python
#GET TMZ ARTICLES USING CURL
with open('combined_links.txt', 'r') as f:
    links = [link.strip() for link in f.readlines()]
    
    
def get_article(link):
    html = !curl -s {link}
    html = ''.join(html)
    soup = BeautifulSoup(html, 'lxml')
    article = ''
    for thing in soup.find_all(class_ = "all-post-body group article-content"):
        for i,thing2 in enumerate(thing.find_all('p')):
            if i < len(thing.find_all('p'))-1:
                article = article + thing2.text.strip()

    article = article.replace('\n', ' ')
    return article

```

```python
with open('/Users/shareshianl/.twitter_config') as file:
    json_data = json.load(file)
oauth = OAuth1(json_data["consumer_key"],
               json_data["consumer_secret"],
               json_data["access_token"],
               json_data["access_token_secret"])
```

```python
tweet_list=[]
for account in accounts:
    TWITTER_URL = 'https://api.twitter.com/1.1/statuses/user_timeline.json'
    acct = account
    print(acct)
    twittercount = 200
    count = 0
    
    full_url=''.join([TWITTER_URL,'?','count=',str(twittercount),'&', 'screen_name=', acct])
    response = requests.get(full_url,auth=oauth)    
    js = json.loads(response.text)
    for i in range(len(js)):
        tweet_list.append([js[i]['user']['screen_name'], js[i]['text']])            

    max_id=js[len(js) - 1]['id']

    while count < 5:
        try: 
            full_url=''.join([TWITTER_URL,'?','count=',str(twittercount),'&', 'screen_name=', acct, '&', 'max_id=', str(max_id)])
            response = requests.get(full_url,auth=oauth)    
            js = json.loads(response.text)
            for i in range(len(js)):
                tweet_list.append([js[i]['user']['screen_name'], js[i]['text']])    
                
            max_id=js[len(js) - 1]['id']
            count = count +1 
        except:
            break
        
```
